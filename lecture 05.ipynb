### Механическая работа

У нас есть интуитивное понятие работы. Работы бывает много или мало, работа бывает легкая или тяжелая. Как говорится, работа не волк - в лес не убежит. В физике для работы, как и для других понятий вводится строгое определение.

Механической работой силы ${\bf F}$ на перемещении $d{\bf r}$ называется величина
$$\delta A = {\bf F}\cdot d{\bf r} = F\left|d{\bf r}\right|\cos(\angle({\bf F},d{\bf r}))$$
Здесь для бесконечно малого приращения работы использован значёк $\delta$, который показывает, что работа, вообще говоря, есть функция процесса - она зависит от пути перехода из начальной точки в конечную. Полная работа записывается в виде интеграла
$$A = \int_{L}{\bf F}\cdot d{\bf r}$$
Так как этот интеграл берется вдоль траектории движения материальной точки, он называется криволинейным, а работа – это работа силы вдоль заданной кривой $L$. Сила здесь должна рассматриваться как суперпозиция сил, действующих на материальную точку. Поэтому работу можно разложить на сумму работ каждой силы в отдельности:
$${\bf F} = \sum_{i}{\bf F}_{i}\Rightarrow A = \int_{L}{\bf F}\cdot d{\bf r} = \int_{L}\sum_{i}{\bf F}_{i}\cdot d{\bf r} = \sum_{i}\int_{L}{\bf F}_{i}\cdot d{\bf r} = \sum_{i}A_{i}$$

При движении по одной и той же траектории из положения 1 в положение 2, а зтем обратно из положения 2 в положение 1 при условии неизменности сил, действующих на материальную точку, работы по этим двум перемещениям отличаются знаком:
$$A_{12}=-A_{21}$$

Часто в приложения важна не только работа, но время, за которое она выполняется: как быстро разгонится автомобиль, или поднимется груз. Скорость выполнения работы называется мощностью - это работа в единицу времени:
$$P=\dfrac{dA}{dt}$$
В системе единиц СИ работу измеряют в джоулях: 1 Дж – это работа силы 1 Н по перемещению тела на расстояние 1 м. Мощность измеряют в Ваттах: 1 Вт – это мощность работы 1 Дж за одну секунду.

Криволинейный интеграл, которым определяется работа, можно рассчитывать по-разному в разных задачах. Если изучается работа консервативных сил, которые обсуждаются ниже, их работа не зависит от траектории, а зависит только от начального и конечнго положения. Сложнее обстоит дело, например, с силой трения. Рассмотрим такой пример: какую минимальную работу нужно затратить, чтобы втащить тело массой $m$ на горку с длиной основания $l$ и высотой $h$, профиль которой задается гладкой неубывающей функией, если коэффициент трения между телом и поверхностью горки равен $\mu$?

При втаскивании совершается работа против сил трения и силы тяжести. Для минимизации работы необходимо обеспечить нулевое ускорение тела в течение всего перемещения. Проецируюя силы на касательное и нормальное направления в данной точке горки, имеем,
$$F = \mu mg\cos\alpha + mg\sin\alpha=\mu$$
Рассмотрим бесконечно малое перемещение вдоль горки $ds$. Его можно спроецировать на декартовы оси координат, используя угол наклона в данной точке:
$$dx = ds\cos\alpha,\;dy = ds\sin\alpha$$
Тогда для элементарной работы
$$dA = Fds = \mu mgds\cos\alpha + mgds\sin\alpha = \mu mgdx + mgdy$$
Интегрирование вдоль поверхности горки сведется к интегрированию по горизонтальной координате первого слагаемого и по вертикальной координате второго:
$$A = \intop_{1}^{2}Fds = \intop_{1}^{2}\left(\mu mgdx+mgdy\right) = \mu mgl + mgh$$
Результат может показаться неожиданным, но он показывает, что для сил, работа которых зависит от траектории, нужно представлять сооветствующий криволинейный интеграл в соответствии с данными и геометрией задачи.

### Кинетическая энергия

Рассмотрим работу сил по перемещению материальной точки и применим второй закон Ньютона, а также учтем, что $d{\bf r} = {\bf v}dt$:
$$A = \int_{L}{\bf F}\cdot d{\bf r} = \int_{L}\dfrac{d{\bf p}}{dt}\cdot{\bf v}dt = \int_{L}d{\bf p}\cdot{\bf v}=m\int_{L}{\bf v}\cdot d{\bf v}$$
Чтобы переписать выражение под интегралом, рассмотрим определение модуля вектора скорости: $v^{2}\equiv{\bf v}\cdot{\bf v}$. Продифференцируем это выражение: $2vdv=d{\bf v}\cdot{\bf v}+{\bf v}\cdot d{\bf v}=2{\bf v}\cdot d{\bf v}$. Следовательно, ${\bf v}\cdot d{\bf v}=vdv$. Если за какой-то промежуток времени при движении вдоль траектории скорость изменилась от $v_{1}$ до $v_{2}$, то
$$A = m\intop_{v_{1}}^{v_{2}}vdv = \left.\dfrac{1}{2}mv^{2}\right|_{v_{1}}^{v_{2}} = \dfrac{1}{2}mv_{2}^{2}-\dfrac{1}{2}mv_{1}^{2} = T_{2}-T_{1}$$
Здесь введена величина $T = mv^{2}/2$ – кинетическая энергия материальной точки. Итак, суммарная работа сил при перемещении материальной точки равна приращению кинетической энергии этой материальной точки. Обобщая на систему материальных точек, можно сказать, что суммарная работа всех сил (и внутренних, и внешних), действующих на систему материальных точек, равна приращению полной кинетической энергии этой системы, которая вычисляется как алгебраическая сумма энергий каждой материальной точки в отдельности.

### Теорема Кёнига

Кинетическая энергия зависит от скорости, а, следовательно, и от выбора системы отсчета. Выясним, как она преобразуется при изменении системы отсчета. Запишем преобразование скорости:
$${\bf v} = {\bf v}'+{\bf V}\Rightarrow v^{2} = {\bf v}\cdot{\bf v} = \left({\bf v}'+{\bf V}\right)\cdot\left({\bf v}'+{\bf V}\right) = v'^{2}+V^{2}+2{\bf v}'\cdot{\bf V}$$
Тогда суммарная кинетическая энергия системы материальных точек
$$K = \sum_{i}\dfrac{1}{2}m_{i}v_{i}^{2} = \sum_{i}\dfrac{1}{2}m_{i}\left(v_{i}'^{2}+2{\bf v}_{i}'\cdot{\bf V}+V^{2}\right) = \sum_{i}\dfrac{1}{2}m_{i}v_{i}'^{2} + \sum_{i}m_{i}{\bf v}_{i}'\cdot{\bf V} + \sum_{i}\dfrac{1}{2}m_{i}V^{2} = K' + \left({\bf P}'\cdot{\bf V}\right)+\dfrac{mV^{2}}{2}$$
Если новая система отсчета - это система отсчета центра масс, то суммарный импульс в этой системе ${\bf P}'={\bf P}_c=0$, и 
$$K = K_c + \dfrac{mV_{c}^{2}}{2}$$
Это соотношение называется теоремой Кёнига: кинетическая энергия системы материальных точек равна сумме кинетической энергии движения этих точек в системе отчета с началом координат в центре масс (СЦМ), которая движется поступательно со скоростью центра масс, и половине произведения всей массы системы на квадра скорости центра масс. Из этой теоремы следует, что кинетическая энергия в СЦМ минимальна.

### Потенциальная энергия

В отношении к производимой работе силы можно разделить на три группы: консервативные и неконсервативные. В случае консервативных сил работа зависит только от начальной и конечной точек траектории, но не от самой траектории. В противоположность, работа неконсервативных сил зависит от траектории. Их еще называют диссипативными, поскольку их работа связана с потерей механической энергии. Иногда еще выделяют гироскопические силы - они всегда перпендикулярны скорости, и их работа тождественно равна нулю. Примером такой силы является магнитная составляющая силы Лоренца.

В случае консервативных сил при перемещении по двум разным траекториям с одинаковыми начальным и конечным положениями, по определению:
$$A_{12}^{(1)} = A_{12}^{(2)}$$
Если изменить направление движения по второй траектории, то $A_{21}^{(2)} = -A_{12}^{(2)}$. Получаем, что $A_{12}^{(1)} + A_{21}^{(2)}=0$. То есть, работа консервативных по любому замкнутому контуру равна нулю.

Работу консервативных сил можно рассматривать как изменение некоторой величины – потенциальной энергии, – которая является функцией только координат. Определим потенциальную энергию как работу при переходе из рассматриваемой точки в некоторе заранее выбранное начало отчета:
$$U_{1}=A_{10}$$
Физически значимой величиной является работа по перемещению тела из начального в конечное положение. Покажем, что при этом точку начала отсчета потенциальной энергии можно выбирать произвольно:
$$U_{1}=A_{10},\;U_{2}=A_{20}\Rightarrow A_{12}=A_{10}+A_{02}=U_{1}-U_{2}$$
Таким образом, работа равна убыли потенциальной энергии. Если при этом рассмотреть два разных начала отсчета энергии:
$$U_{1}=A_{10},\; U_{1}'=A_{10'}\Rightarrow U_{1}=A_{10}=A_{10}+A_{00'}+A_{0'0}=U_{1}'+A_{0'0}\Rightarrow$$
$$\Rightarrow A_{12}=U_{1}-U_{2}=U_{1}'+A_{0'0}-U_{2}'-A_{0'0}=U_{1}'-U_{2}'$$

В качестве первого примера рассмотрим элементарную работу силы тяжести при малом изменении высоты, направив ось $Z$ системы координат по вертикали, заданной направлением ускорения свободного падения: $dA = m{\bf g}\cdot d{\bf r} = mg|{\bf r}|\cos\alpha = -mgdz$. Разбивая произвольную траекторию тела на малые участки по высоте и интегрируя, получаем, что полная работа зависит только от разности высот и не зависит от формы траектории:
$$A_{12} = -\intop_{z_{1}}^{z_{2}}mgdz = mg\left(z_{1}-z_{2}\right) = U_1 - U_2$$
где потенциальная энергия $U = mgz$, а начало ее отсчета совпадает с началом выбранной системы координат.

Другим важным примером является работа центральной силы - силы, которая зависит только расстояния $r$ до некоторого притягивающего центра (примеры - силы кулоновского и гравитационного взаимодействия). Такую силу можно записать как: ${\bf F}=F\left(r\right)\dfrac{{\bf r}}{r}$. Центральная сила является консервативной, поскольку её работа зависит только расстояния до силового центра:
$$dA = {\bf F}\cdot d{\bf r} = F\left(r\right)\dfrac{{\bf r}\cdot d{\bf r}}{r} = Fdr\Rightarrow A=\intop_{r_{1}}^{r_{2}}F\left(r\right)dr=\Phi\left(r_{2}\right)-\Phi\left(r_{1}\right)$$
Здесь учтено, что $r^{2}={\bf r}\cdot{\bf r}\Rightarrow rdr={\bf r}\cdot d{\bf r}$. Потенциальная энергия центральной силы:
$$A = \intop_{r_{1}}^{r_{2}}F\left(r\right)dr = \Phi\left(r_{2}\right)-\Phi\left(r_{1}\right)\Rightarrow U=-\Phi\left(r\right) + const$$
где $\Phi$ - некоторая первообразная функции $F$. В случае гравитационного взаимодействия в качестве точки отсчета потенциальной энергии часто выбирается бесконечно удаленная точка:
$$U(r) = A_{10} = \intop_{r}^{\infty}{\bf F}\left({\bf r}\right)d{\bf r} = -\intop_{r}^{\infty}\gamma\dfrac{m_{1}m_{2}}{r^{2}}\dfrac{{\bf r}}{r}d{\bf r} = -\intop_{r}^{\infty}\gamma\dfrac{m_{1}m_{2}}{r^{2}}\dfrac{rdr}{r} = \left.\gamma\dfrac{m_{1}m_{2}}{r}\right|_{r}^{\infty} = -\gamma\dfrac{m_{1}m_{2}}{r}$$

Консервативные силы представляют собой важный пример, поскольку в микромире все силы так или иначе являются консервативными.

### Полная механическая энергия

Пусть на материальные точки некоторой системы действуют только консервативные силы. Тогда, с одной стороны, работа по перемещению точек равна убыли потенциальной энергии, а, с другой, изменению кинетической энергии. Тогда
$$A_{12} = U_{1}-U_{2} = T_{2}-T_{1}\Rightarrow T+U=E=const$$
То есть, в системе, где действуют исключительно консервативные силы полная механическая энергия остается постоянной. Если же присутствуют еще и диссипативные силы, то работу всех сил можно разделить на работу консервативных и диссипативных сил по отдельности и учесть, что кинетическая энергия изменяется за счет работы всех сил:
$$A_{12} = A_{12}^{к} + A_{12}^{д} = U_{1}-U_{2} + A_{12}^{д} = T_{2}-T_{1}\Rightarrow E_{2}-E_{1} = A_{12}^{д}$$
В данном случае изменение полной энергии равно работе диссипативных сил. Эта работа, как можно видеть, является отрицательной. При наличии диссипативных сил в системе механическая энергия переходит в другие формы, например, тепловую.
