### Cилы в макроскопической динамике

В прошлой лекции мы обсуждали, что в природе существует всего четыре фундаментальные взаимодействия. Тем не менее, в практических задачах зачастую использовать фундаментальные силы, соответствующие этим взаимодействиям, неудобно и даже невозможно, так что взаимодействие физических объектов описывают с помощью эффективных сил. Так, например, сила упругости пружины и сила трения в конечном итоге сводятся к электромагнитным взаиомодействиям на микроуровне, но в макроскопических задачах такие микросилы во многих случаях по отдельности учесть невозможно, даже с учетом имеющихся сейчас вычислительных мощностей. Главное требование к таким эффективным силам - адекватное описание состояний физических систем в рамках используемых моделей (в нашем случае - в рамках моделей классической механики). Рассмотрим примеры проявления силы трения.

### Трение в блоках

В качестве примера проявления и применения силы трения рассмотрим задачу о трении в блоке. Пусть блок поворачивает легкую нерастяжимую веревку на угол $\alpha$, причем между веревкой и блоком действует сила трения, а коэффициент трения равен $\mu$. Найдем, как отличаются силы натяжения веревки с двух сторон от блока.

Рассмотрим малый участок веревки, на котором она поворачивается на малый угол $\Delta\alpha$. На участок действуют силы натяжения $T$ и $T+\Delta T$ со стороны остальной части веревки, сила реакции опоры $N$ со стороны блока, и сила трения $F = \mu N$. Спроектируем уравнение движения на ось, касательную к блоку в данной точке, и ось, перпендикулярную блоку, считая веревку неподвижной:

$$F = (T(\alpha) - T(\alpha+\Delta\alpha))\cos(\Delta\alpha/2) = -\Delta T \cos(\Delta\alpha/2)$$
$$N = (T(\alpha+\Delta\alpha) + T(\alpha))\sin(\Delta\alpha/2)$$

Учтем, что угол $\Delta\alpha$ малый, поэтому $\cos(\Delta\alpha/2)\approx 1$, $\sin(\Delta\alpha/2)\approx\Delta\alpha/2$ и $T(\alpha+\Delta\alpha) + T(\alpha) \approx 2T$. Тогда

$$\mu T\Delta\alpha \approx -\Delta T \Rightarrow \frac{\Delta T}{T} \approx -\mu \Delta\alpha$$

Перейдем к бесконечно малым приращениям и проинтегрируем по углу от нуля до какого-то заданного $\alpha_0$:

$$\int\limits_{T_1}^{T_2}\frac{\Delta T}{T} = -\mu \int\limits_{0}^{\alpha_0}\Delta\alpha \Rightarrow \ln\frac{T_2}{T_1} = -\mu\alpha_0 \Rightarrow T_2 = T_1\exp(-\mu\alpha_0)$$

Последнее выражение иногда называется формулой Эйлера.

### Вязкое трение

При движении в вязкой среде, будь то жидкость или газ, механические объекты испытывают сопротивление. При достаточно медленных движениях сила сопротивления пропорциональна скорости, а при быстрых - квадрату скорости тела. Переход между этими режимами определяется числом Рейнольдса. Физически разница между режимами определяется относительной ролью силы трения между телом и примыкающими к нему слоями вязкой среды и силы динамического сопротивления, возникающей при столкновении движущегося тела с неподвижными массами жидкости или газа. Если преобладает первая сила, то движение называется медленным, если вторая - быстрым.

Рассмотрим пример задачи, где возможно найти аналитическое решение уравнений движения тела в вязкой среде. Пусть снаряд запущен с начальной скоростью $700$ м/с под углом $30^{\circ}$ к горизонту. Будем считать, что сила сопротивления пропорциональна скорости ${\bf F}=-k{\bf v}$. При этом известно, что при падении снаряда с большой высоты его установившаяся скорость равна $v_{\infty} = 200$ м/с.

Сначала рассмотрим свободное падение снаряда. Проекция уравнения движения на вертикальную ось дает:

$$m\frac{dv_y}{dt} = mg - kv_y \Rightarrow \frac{m}{k}\frac{dv_y}{mg/k - v_y} = dt \Rightarrow \frac{m}{k}\int\limits_{0}^{v_y(t)}\dfrac{dv_y}{mg/k - v_y} = \int\limits_{0}^{t}dt $$

$$\Rightarrow v_y(t) = \frac{mg}{k}\left[1-\exp\left(-\frac{k}{m}t\right)\right]$$

Данная в условии установившая скорость - это скорость при $t\rightarrow\infty$. Получаем, что

$$v_{\infty} = \frac{mg}{k}$$

Теперь вернемся к задаче и спроецируем уравнение движения на вертикальную и горизонтальную оси координат:

$$m\dfrac{d{\bf v}}{dt}=m{\bf g}-k{\bf v}\Rightarrow\left\{ \begin{array}{c}
m\dfrac{dv_{x}}{dt}=-kv_{x}\\
m\dfrac{dv_{y}}{dt}=-mg-kv_{y}
\end{array}\right.$$

Эти уравнения можно решать раздельно. Для движения по горизонтали

$$\frac{m}{k}\frac{dv_x}{v_x} = -dt \Rightarrow \frac{m}{k}\int\limits_{v_{x0}}^{v_x}\frac{dv_x}{v_x} = -\int\limits_{0}^{t}dt \Rightarrow \ln \frac{v_x(t)}{v_{x0}} = -\frac{k}{m}t \Rightarrow v_x(t) = v_{x0}\exp\left(-\frac{k}{m}t\right)$$

Зависимость горизонтальной координаты от времени:

$$\frac{dx}{dt} = v_{x0}\exp\left(-\dfrac{k}{m}t\right) \Rightarrow \int\limits_{0}^{x} dx = v_{x0}\int\limits_{0}^{t}\exp\left(-\dfrac{k}{m}t\right)dt \Rightarrow x\left(t\right)=\dfrac{mv_{0x}}{k}\left[1-\exp\left(-\dfrac{k}{m}t\right)\right] $$

Для движения по вертикали

$$m\dfrac{dv_{y}}{dt}=-mg-kv_{y} \Rightarrow \frac{m}{k}\intop_{v_{y0}}^{v_{y}}\dfrac{dv_y}{mg/k+v_y}=-\intop_{0}^{t}dt \Rightarrow v_{y}\left(t\right)=\left(v_{y0} + v_{\infty}\right)\exp\left(-\dfrac{k}{m}t\right)-v_{\infty}$$

Здесь мы заменили полученное выражение для скорости $v_{\infty}$. Зависимость вертикальной координаты от времени:

$$\dfrac{dy}{dt}=\left(v_{y0} + v_{\infty}\right)\exp\left(-\dfrac{k}{m}t\right)-v_{\infty} \Rightarrow$$

$$\Rightarrow \intop_{0}^{y}dy=\intop_{0}^{t}\left[\left(v_{y0}+v_{\infty}\right)\exp\left(-\dfrac{k}{m}t\right)-v_{\infty}\right]dt \Rightarrow y\left(t\right)=\dfrac{m}{k}\left(v_{y0}+v_{\infty}\right)\left[1-\exp\left(-\dfrac{k}{m}t\right)\right] - v_{\infty}t $$

Легко проверить, что при $k\rightarrow 0$ получается известный кинематический результат - движение по параболе.

Найдем максимальную высоту подъема и расстояние по горизонтали до этой точки из условия $v_{y}=0$:

$$\left(v_{y0}+v_{\infty}\right)\exp\left(-\dfrac{k}{m}t\right) = v_{\infty}\Rightarrow t_{H} = \dfrac{m}{k}\ln\left(1 + \dfrac{v_{y0}}{v_{\infty}}\right)$$

$$\Rightarrow H=y\left(t_{H}\right)=\dfrac{m}{k}\left[v_{y0}-v_{\infty}\ln\left(1+\dfrac{v_{y0}}{v_{\infty}}\right)\right]=\dfrac{v_{\infty}^{2}}{g}\left[\dfrac{v_{0}\sin\alpha}{v_{\infty}}-\ln\left(1+\dfrac{v_{0}\sin\alpha}{v_{\infty}}\right)\right]\approx3.01\text{км}$$

Расстояние до точки максимального подъема:

$$x\left(t_{H}\right)=\dfrac{v_{0}^{2}\sin2\alpha}{2g}\left(1+\dfrac{v_{0}\sin\alpha}{v_{\infty}}\right)^{-1}\approx7.87\text{км}$$

```python
# Рассмотрим решение задачи с помощью библиотеки символьных вычислений sympy:

from sympy import *
from sympy.solvers.ode.ode import constantsimp
from sympy.plotting import plot_parametric

## объявление параметров и функций:

# параметры задачи: масса, коэффициент вязкого трения,
#   ускорение свободного падения, время,
#   проекции начальной скорости, модуль начальной скорости и угол выстрела
m, k, g, t, vx0, vy0, v0, ang = symbols('m k g t vx0 vy0 v0 ang', positive=True)
# неизвестные функции времени: координаты и проекции скорости
x, y, vx, vy = symbols('x y vx vy', cls=Function)

## решение уравнений движения:

# уравнение для производной горизонтальной компоненты скорости:
eq_vx = Eq(vx(t).diff(t), ( -(k/m) * vx(t)))
# находим зависимость горизонтальной компоненты скорости от времени:
vx_res = dsolve(eq_vx, ics={vx(0): vx0})
print('зависимость горизонтальной компоненты скорости от времени:')
pprint(vx_res)
# уравнение для производной горизонтальной компоненты координаты:
eq_x = Eq(x(t).diff(t), vx_res.rhs)
# находим зависимость горизонтальной компоненты координаты от времени:
x_res = dsolve(eq_x, ics={x(0): 0})
print('зависимость горизонтальной координаты от времени:')
pprint(x_res)

# уравнение для производной вертикальной компоненты скорости:
eq_vy = Eq(vy(t).diff(t), (-g - (k/m) * vy(t)))
# находим зависимость вертикальной компоненты скорости от времени:
vy_res = dsolve(eq_vy, ics={vy(0): vy0})
print('зависимость вертикальной компоненты скорости от времени:')
pprint(vy_res)
# уравнение для производной вертикальной компоненты координаты:
eq_y = Eq(y(t).diff(t), vy_res.rhs)
# находим зависимость вертикальной компоненты координаты от времени:
y_res = dsolve(eq_y, ics={y(0): 0})
print('зависимость вертикальной координаты от времени:')
pprint(y_res)

## численные результаты:

# численные значения параметров:
ang = 30 # угол выстрела
v0 = 700 # модуль начальной скорости снаряда
substitution = [(m, 1), (k, 0.049), (g, 9.8), (vx0, v0*cos(ang*pi/180)), (vy0, v0*sin(ang*pi/180))]

# аналитические и численные результаты без учета сопротивления воздуха:
x_res_free = vx0*t
y_res_free = vy0*t - 0.5*g*t**2
x_fun_free = x_res_free.subs(substitution)
y_fun_free = y_res_free.subs(substitution)
    # время полета:
ft_max_free = 2*vy0/g
t_max_free = ft_max_free.subs(substitution);
print('время полета без учета сопротивления воздуха: %.2fс' % t_max_free)
    # максимальная высота:
h_max_free = 0.001*y_fun_free.evalf().subs(t, 0.5*t_max_free)
print('максимальная высота без учета сопротивления воздуха: %.2f км' % h_max_free)
    # дальность полета:
l_max_free = 0.001*x_fun_free.evalf().subs(t, t_max_free)
print('дальность полета без учета сопротивления воздуха: %.2f км' % l_max_free)

# аналитические и численные результаты с учетом сопротивления воздуха:
x_fun = x_res.rhs.subs(substitution)
y_fun = y_res.rhs.subs(substitution)
    # найдем время полета через условие x(t)=0 с начальным условием, заданным t_max_free:
t_max = nsolve(y_fun, t, t_max_free)
print('время полета с учетом сопротивления воздуха: %.2fс' % t_max)
    # найдем время полета до максимальной высоты через условие vy(t)=0:
t_max_half = nsolve(vy_res.rhs.subs(substitution), t, 0.5*t_max_free)
print('время полета до максимальной высоты с учетом сопротивления воздуха: %.2fс' % t_max_half)
    # максимальная высота:
h_max = 0.001*y_fun.evalf().subs(t, t_max_half)
print('максимальная высота с учетом сопротивления воздуха: %.2f км' % h_max)
    # дальность полета:
l_max = 0.001*x_fun.evalf().subs(t, t_max)
print('дальность полета с учетом сопротивления воздуха: %.2f км' % l_max)

## графики:
import matplotlib.pyplot as plt
# построим траекторию без учета сопротивления воздуха (красная) и с учетом сопротивления (синяя):
p = plot_parametric((x_fun, y_fun, (t, 0, t_max)), 
                    (x_fun_free, y_fun_free, (t, 0, t_max_free)), 
                    show=False)
x1,y1 = p[0].get_points()
x2,y2 = p[1].get_points()
plt.plot(0.001*x1, 0.001*y1, label = 'с сопротивлением воздуха')
plt.plot(0.001*x2, 0.001*y2, label = 'без сопротивления воздуха')
plt.legend()
plt.xlim(0)
plt.ylim(0)
plt.xlabel('x, км')
plt.ylabel('y, км')
plt.show()
```

### Реактивное движение

Рассмотрим движение с переменной массой – движение при котором часть масса рассматриваемого механического объекта изменяется, что приводит к изменению импульса этого объекта. Частным случаем является реактивное движением, при котором масса ракеты выбрасывается для ускорения оставшейся части. Для краткости рассмотрим ракету и вылетающий из нее газ, и запишем изменение импульса системы «ракета – выхлопные газы» за малый промежуток времени $dt$:

$$\left(m+dm\right)\left({\bf v}+d{\bf v}\right)+dm_{gas}{\bf v}_{gas}-m{\bf v}={\bf F}dt$$

Раскроем скобки и пренебрежем малой величиной второго порядка $dmd{\bf v}$, а также введем скорость вылета газов относительно ракеты ${\bf u}={\bf v}_{gas}-{\bf v}$. В силу закона сохранения массы $m+m_{gas}=const\Rightarrow dm=-dm_{gas}$. Тогда

$$md{\bf v}+dm_{gas}{\bf u}_{gas}={\bf F}dt\Rightarrow m\dfrac{d{\bf v}}{dt}=\dfrac{dm}{dt}{\bf u}+{\bf F}$$

Это уравнение называется уравнением Мещерского. Оно показывает, что, кроме внешней силы, на тело действует эффективная реактивная сила $\dfrac{dm}{dt}{\bf u}$.

Теперь для простоты предположим, что ракета движется прямолинейно, и на нее не действуют внешние силы. Поскольку скорость газа направлена в противоположном направлении движению ракеты, то проекция уравнения на направление движения будет

$$m\dfrac{dv}{dt}=-\dfrac{dm}{dt}u\Rightarrow dv=-u\dfrac{dm}{m}\Rightarrow\intop_{v_{1}}^{v_{2}}dv=-u\intop_{m_{1}}^{m_{2}}\dfrac{dm}{m}\Rightarrow\dfrac{v_{2}-v_{1}}{u}=\ln\dfrac{m_{1}}{m_{2}}$$

Последнее выражение называется формулой Циолковского.

Рассмотрим пример задачи с переменной массой: две одинаковые легкие тележки, на которых сидят два одинаковых дворника, катятся по инерции прямолинейно параллельно друг другу с одинаковыми скоростями по очень скользкому льду. Начинает падать снег. Дворник, сидящий на одной из тележек, сбрасывает падающий снег с тележки в стороны перпендикулярно движению, а дворник на второй тележке спит. Какая из тележек быстрее проедет одно и то же расстояние?

Если рассуждать качественно, можно приблизительно записать закон сохранения импульса в проекции на направление движения тележек: $mv_{1}=\left(m+\Delta m\right)v_{2}$. Видно, что уменьшение скорости определяется отношением $m/\left(m+\Delta m\right)$. Для тележки, с которой сметается снег, оно постоянно, а для тележки, на которой накапливается снег, оно уменьшается. Поэтому вторая будет тормозиться медленнее. Чтобы решить задачу точнее, запишем уравнение движения (воспользуемся уравнение Мещерского):

$$m\dfrac{dv}{dt}=-v\dfrac{dm}{dt}$$

В случае спящего дворника:

$$m\dfrac{dv}{dt}=-\mu v\Rightarrow\dfrac{dv}{v}=-\mu\dfrac{dt}{m_{0}+\mu t}\Rightarrow\ln\dfrac{v\left(t\right)}{v_{0}}=-\ln\dfrac{m_{0}+\mu t}{m_{0}}\Rightarrow v\left(t\right)=v_{0}\dfrac{m_{0}}{m_{0}+\mu t}$$

В случае работающего дворника:

$$m_{0}\dfrac{dv}{dt}=-\mu v\Rightarrow\dfrac{dv}{v}=-\dfrac{\mu}{m}dt\Rightarrow v=v_{0}\exp\left(-\dfrac{\mu}{m}dt\right)$$

Экспоненциальная функция убывает быстрее стеменной, что подтверждает сделанное качественное предположение.
