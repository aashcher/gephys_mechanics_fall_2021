### Вращательное движение

На предыдущих лекциях мы рассматривали поступательное движение материальной точки. Динамика поступательного движения описывается вторым законом Ньютона $d{\bf p}/dt={\bf F}$. Для изучения вращательного движения оказывается удобно сформулировать другое динамическое уравнение. Чтобы показать, какие соображения к нему приводят, посмотрим на материальную точку, которая движется по окружности радиуса $R$, и на которую действует сила $F$, все время направленная по касательной к окружности. Движение такой точки удобно характеризовать с помощью угловой скорости $\omega = v/R$, поэтому из второго закона Ньтона следует:
$$m\frac{dv}{dt} = F \Rightarrow mR^2\frac{d\omega}{dt}=FR \Rightarrow \frac{d(mR^2\omega)}{dt}=FR$$
Величина, стоящая в правой части есть момент силы $F$ относительно центра окружности. Величину, стоящую под знаком дифференцирования, можно переписать как $L = mR^2\omega = mvR = pR$. Будем называть ее моментом импульса.

### Векторное произведение

Чтобы обобщить понятия момента силы и момента импульса и ввести их в виде векторных величин, рассмотрим необходимые математические определения. Векторным проиведением двух векторов $\bf a$ и $\bf b$ называется вектор
$${\bf c} = {\bf a}\times{\bf b}$$
такой, что ${\bf c}\perp{\bf a}$, ${\bf c}\perp{\bf b}$, направление вектора ${\bf c}$ определяется направлением правого винта при кратчайшем повороте от $\bf a$ к $\bf b$, и $|{\bf c}|=|{\bf a}||{\bf b}|\sin(\angle({\bf a},{\bf b}))$. Геометрически модуль вектрного произведения равен площади параллелограмма, построенного на векторах $\bf a$ и $\bf b$, а его направление ортогонально плоскости этого параллелограмма. В декартовых координатах с единичными ортами $\hat{\bf e}_x$, $\hat{\bf e}_y$, $\hat{\bf e}_z$ векторное произведение можно записать через детерминант
$${\bf c} = \left|\begin{array}.\hat{\bf e}_x&\hat{\bf e}_y&\hat{\bf e}_z\\a_x&a_y&a_z\\b_x&b_y&b_z\end{array}\right| = \hat{\bf e}_x(a_yb_z-a_zb_y) + \hat{\bf e}_y(a_zb_x-a_xb_z) + \hat{\bf e}_z(a_xb_y-a_yb_x)$$
Из определения вектроного произведения следует, что, если ${\bf a} || {\bf b}$, то ${\bf a}\times{\bf b}=0$. Можно также встретить и другие обозначения вектрного произвдеения: ${\bf c} = [{\bf a},{\bf b}]$ или ${\bf c} = {\bf a}\wedge{\bf b}$.

### Момент силы

Теперь можно обощить определения моментов силы и импульса и ввести и как векторные величины. Моментом силы $\bf F$ называется вектор
$${\bf M} = {\bf r}\times{\bf F}$$
Если рассмотреть прямую, проходящую через точку приложения силы и параллельную вектору силы, - линию действия силы, - то из определения векторного произведения следует, что $M=Fl$, где $l$ - кратчайшее расстояние от начала координат до линии действия силы, $l=|{\bf r}|\sin(\angle({\bf F},{\bf r}))$. Видно, что, если переносить точку приложения силы вдоль линии действия, момент не изменится.

Чтобы найти момент сил, действующийх на систему материальных точек, нужно сложить все моменты:
$${\bf M} = \sum_i{\bf M}_i = \sum_i{\bf r}_i\times{\bf F}_i$$
Поскольку, по определению, момент зависит от выбора начала отсчета, то при переносе этого начала, ${\bf r}_i = {\bf r}_i' + {\bf R}$, и
$${\bf M} = \sum_i({\bf r}_i' + {\bf R})\times{\bf F}_i = \sum_i{\bf r}_i'\times{\bf F}_i + {\bf R}\times\sum_i{\bf F}_i = {\bf M}' + {\bf R}\times{\bf F}$$
где ${\bf F}$ - равнодействующая действующих на систему сил.

Важным частным случаем является так называемый момент пары сил - суммарный момент двух сил, равных по модулю, противоположных по направлению, и приложенных к разным точкам (${\bf F}_1 = -{\bf F}_2$). По определению,
$${\bf M} = {\bf r}_1\times{\bf F}_1 + {\bf r}_2\times{\bf F}_2 = ({\bf r}_1 - {\bf r}_2)\times{\bf F}_1$$
то есть, момент пары сил не зависит от начала отсчета и определяется только расстоянием между линиями действия сил.

### Момент импульса

Моментом импульса материальной точки, имеющей импульс $\bf p$, называется вектор
$${\bf L} = {\bf r}\times{\bf p}$$
В случае системы материальных точек момент будет векторной суммой моментов каждой точки:
$${\bf L} = \sum_i{\bf r}_i\times{\bf p}_i$$

Рассмотрим изменение момента ${\bf L}$ системы материальных точек при переходе в новую систему отсчета, ${\bf r}_i={\bf r}'_i+{\bf R}$, ${\bf v}_i={\bf v}'_i+{\bf V}$,
$${\bf L} = \sum_i{\bf r}_i\times{\bf p}_i = \sum_i m_i \left({\bf r}'_i+{\bf R}\right)\times\left({\bf v}'_i+{\bf V}\right) =$$
$$= \sum_i m_i {\bf r}'_i \times {\bf v}'_i + \left(\sum_i m_i {\bf r}'_i\right) \times {\bf V} + {\bf R} \times \left(\sum_i m_i  {\bf v}'_i\right) + \left(\sum_i m_i\right) {\bf R} \times {\bf V}$$
Здесь мы воспользовались линейностью векторного произведения и вынесли за скобки во втрой и третьей сумме одинаковые сомножители. Заметим, что первая сумма есть, по определению, момент импульса в новой системе отсчета, сумма во втором слагаемом - полная масса системы, умноженная на радиус-вектор центра масс в новой системе отсчета ${\bf R}'_c$, а сумма в третьем слагаемом - полный импульс в новой системе отсчета ${\bf p}'$:
$${\bf L} = {\bf L}' + m {\bf R}'_c \times {\bf V} + {\bf R} \times {\bf p}' + m {\bf R} \times {\bf V}$$
Если новая система отсчета - система отсчета центра масс, то ${\bf R}'_c\equiv0$, ${\bf p}'=0$, и, следовательно, в этом частном случае
$${\bf L} = {\bf L}_с + m {\bf R}_с \times {\bf V}_с$$

### Уравнение моментов и закон сохранения момента импульса

Рассмотрим изменение момента импульса материальной точки во времени:
$$\dfrac{d{\bf L}}{dt}=\dfrac{d{\bf r}}{dt}\times{\bf p}+{\bf r}\times\dfrac{d{\bf p}}{dt} = {\bf r}\times\dfrac{d{\bf p}}{dt}={\bf r}\times{\bf F}={\bf M}$$
Первое слагаемое здесь тождественно равно нулю по свойству векторного произведения, поскольку $d{\bf r}/dt={\bf v}||{\bf p}\Rightarrow {\bf v}\times{\bf p}\equiv 0$. Таким образом, производная момента импульса материальной точки относительно некоторого неподвижного начала $O$ равна моменту сил, действующих на эту точку относительно того же начала. Для системы, состоящей из $N$ материальных, точек имеем:
$$\dfrac{d{\bf L}}{dt}=\sum_{k=1}^{N}\dfrac{d{\bf L}_{k}}{dt}=\sum_{k=1}^{N}\left(\dfrac{d{\bf r}_{k}}{dt}\times{\bf p}_{k}+{\bf r}_{k}\times\dfrac{d{\bf p}_{k}}{dt}\right) = \sum_{k=1}^{N}{\bf r}_{k}\times\dfrac{d{\bf p}_{k}}{dt} $$
По второму закону Ньютона, производная импульса $k$-й материальной точки есть суммарная действующая на нее сила. Представим эту силу в виде суммы некотрой внешней по отношению системе силы ${\bf F}_{ext,k}$ и сил, дествующих на эту точку со стороны всех остальных $N-1$ точек (подчиняющихся третьему закону Ньютона):
$$\dfrac{d{\bf L}}{dt}=\sum_{k=1}^{N}{\bf r}_{k}\times\left({\bf F}_{ext,k}+\sum_{i=1,i\neq k}^{N}{\bf F}_{ki}\right) = $$
$$=\sum_{k=1}^{N}{\bf r}_{k}\times{\bf F}_{ext,k}+\sum_{k=1}^{N-1}\sum_{i=k+1}^{N}\left({\bf r}_{k}-{\bf r}_{i}\right)\times{\bf F}_{ki}=\sum_{k}{\bf r}_{k}\times{\bf F}_{ext,k}={\bf M}_{ext}$$
При раскрытии скобок было учтено, что в двойной сумме каждое слагаемое заданными индексами $i$ и $k$ встречается дважды, и по третьему закон Ньютона, ${\bf F}_{ik} = -{\bf F}_{ki}$.

Полученное уравнение называется уравнением моментов. Оно показывает, что суммарный момент импульса системы материальных точек изменяется только под действием суммарного момента внешних по отношению к системе сил. Этот закон аналогичен втрому закону Ньютона для системы материальных точек.

Отсюда следует закон сохранения момента импульса (ЗСМИ): если на систему материальных точек не действуют внешние силы, или векторная сумма их моментов равна нулю, момент импульса этой системы сохраняется. Закон сохранения момента импулься являтся одним из основных законов физики наравне с законами сохранения энергии и импльса. В курсе теоретической механики будет показано, что ЗСМИ связан с фундаментальным свойством пространства - изотропностью, которое подразумевает, что все направления в пространстве эквивалентны с точки зрения выполнения физических законов.

Важным частным случаем в механике, когда момент импульса сохраняется, является движение в центральном поле. Это движение материальной точки в силовом поле некоторого притягивающего или отталкивающего центра, так что сила зависит от расстояния до этого центра как
$${\bf F} = F(r)\frac{{\bf r}}{r},\;r=|{\bf r}|$$
Момент такой силы является тождественным нулем в системе отсчета, начало которой совпадает с силовым центром, следовательно, в этой системе сохраняется и момент импульса материальной точки. Поскольку момент импульса сохраняется, движение будет плоским, то есть, будет происходить исключительно в плоскости, перпендикулярной вектору $\bf L$.

### Момент импульса относительно оси. Момент инерции.

Возвращаясь к примеру, с которого мы начали лекцию, запишем динамику вращения относительно некоторой фиксированной оси $z$ уже проецируя общее векторное уравнение. Разложим радиус-вектор, векторы импульса и силы на составляющие, параллельные и перпендикулярные этой оси. Тогда проекция момента импульса на 
$$L_{z}=\left({\bf r}\times{\bf p}\right)_{z}=\left[\left({\bf r}_{||}+{\bf r}_{\perp}\right)\times\left({\bf p}_{||}+{\bf p}_{\perp}\right)\right]_{z}=\left({\bf r}_{\perp}\times{\bf p}_{\perp}\right)_{z}$$
При вращении материальной точки массы $m$ вокруг оси z $R=const$, $v=\omega R$, $p=mv=v\omega R$. Тогда проекция уравнения моментов на ось вращения
$$\dfrac{dL_{z}}{dt}=M_{z}=RF_{\perp}\sin\varphi_{F}=lF_{\perp}\Rightarrow mR^{2}\dfrac{d\omega}{dt}=lF_{\perp}$$
где $l$ - плечо силы. Если сопоставить полученное уравнение с вторым законом Ньютона, можно видеть, что роль меры инерции вращательного движения, которую играет масса при поступательном движении, играет величина $I=mR^2$. Эта величина называется моментом инерции. При движении системы материальных точек вокруг оси момент инерции будет суммой $I=\sum_im_iR_i^2$.

### Рассеяние Резерфорда

Практически важным примером применения ЗСМИ является движение в центральном поле, например, в случае гравитационного или кулоновского взаимодействия. Гравитационную задачу двух тел мы рассмотрим на следующих занятиях. Сейчас приведем пример рассеяния альфа-частиц (ядра атома гелия) на ядрах атомов золота. С помощью экспериментального наблюдения такого рассеяния Э. Резерфорд в 1911 году смог понять строение атома и сделать выбор между имеющимися на тот момент различными моделями строения атома. Чтобы описать этот эксперимент оказыватся достаточно простой механической модели, где рассматривается кулоновское взаимодействие двух частиц - изначально неподвижной мишени (ядра атома золота) массы $M$ и налетающей из бесконечности альфа-частицы массы $m$.

Чтобы записать законы сохранения удобным образом, перейдем в полярную систему координат $(r,\varphi)$, так что $x=r\cos\varphi$, $y=r\sin\varphi$. Единичные орты $\hat{\bf e}_r$ и $\hat{\bf e}_{\varphi}$в этой системе координат в каждой точке направлены вдоль радиус-вектора и перпендикулярно ему в направлении возрастания угла $\varphi$ соответственно, как показано на рисунке. Скорость, как и любой вектор, может быть разложена по этим двум ортам: 
$${\bf v} = \frac{d{\bf r}}{dt} = {\bf v}_r + {\bf v}_{\varphi} = v_r\hat{\bf e}_r + v_{\varphi}\hat{\bf e}_{\varphi}.$$
Тогда, поскольку $d\varphi/dt$ - угловая скорость вращения радиус-вектора, то $v_r=dr/dt$ и $v_{\varphi}=rd\varphi/dt$.

Согласно ранее полученному результату для задачи двух тел, при рассеянии Резерфорда можно рассматривать движение одной налетающей частицы с приведенной массой $\mu=mM/\left(m+M\right)$ и решать задачу на относительное расстояние $r$ и относительную скорость $v$. Пусть на бесконечном удалении скорость налетающей частицы равна $v_{\infty}$. Введем так называемый прицельный параметр $b$ - расстояние между параллельными прямыми, одна из которых проходит на бесконечности через налетающую частицу и направлена вдоль ее скорости, а вторая - через частицу-мишень. Запишем импульс налетающей частицы и воспользуемся ЗСМИ:
$${\bf L} = {\bf r}\times\mu{\bf v} = {\bf r}\times\mu{\bf v}_{\varphi} = const$$
поскольку ${\bf v}_r||{\bf r}$. Вектор ${\bf L}$ перпендикулярен плоскости движения и в скалярном виде для некоторого промежуточного положения налетающей частицы и в случае ее бесконечного удаления:
$$L = \mu rv_{\varphi} = \mu bv_{\infty} \Rightarrow \mu r^2\frac{d\varphi}{dt} = \mu bv_{\infty}$$
Пусть $\theta_0$ - угол рассеяния (отклонения налетающей частицы от первначального направления движения) в рассматриваемой системе отсчета, где мишень неподвижна. Полное изменение импульса частицы будет $\Delta p = 2p_0\sin(\theta_0/2)$. Это же изменение можно посчитать с помощью интегрирования импульса силы, спроецировав его на направление $\Delta{\bf p}$. Учитывая, что закон Кулона имеет вид $F = C/r^2$, получаем:
$$\Delta{\bf p}=\intop_{0}^{\infty}{\bf F}dt\Rightarrow\Delta p=\intop_{0}^{\infty}F\cos\psi dt=\intop_{0}^{\pi-\theta_{0}}\dfrac{C}{r^{2}}\cos\left(\dfrac{\pi}{2}-\varphi-\dfrac{\theta_{0}}{2}\right)\dfrac{r^{2}d\varphi}{v_{\infty}b} =$$
$$=\dfrac{C}{v_{\infty}b}\intop_{0}^{\pi-\theta_{0}}\sin\left(\varphi+\dfrac{\theta_{0}}{2}\right)d\varphi=\dfrac{2C}{v_{\infty}b}\cos\dfrac{\theta_{0}}{2}$$
$$\Rightarrow2\mu v_{\infty}\sin\dfrac{\theta}{2}=\dfrac{2C}{v_{\infty}b}\cos\dfrac{\theta_{0}}{2}\Rightarrow\tan\dfrac{\theta_{0}}{2}=\dfrac{C}{\mu v_{\infty}^{2}b}$$
Таким образом, нам удалось связать начальную скорость и прицельный параметр с углом рассеяния.
