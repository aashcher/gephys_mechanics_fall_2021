### Постулаты СТО и относительность одновременности


Специальная теория относительности (СТО) - теория движения со скоростями, близкими к скорости света. СТО как единая теория была сформулирована А. Эйнштейном на основании двух постулатов:
* принцип относительности: все физические явления протекают по одним и тем же законам во всех инерциальных системах отсчета
* скорость света в вакууме однакова во всех инерацильных системах отсчета

В настоящее время второй постулат понимается в более широком смысле: скорость света не только не зависит от системы отсчета, но и является предельной скоростью распространения информационных сигналов.

Несовместимость классической механики, основанной на преобразовании Галилея, с указанными постулатами сразу следует из закона преобразования скоростей, который не может обеспечить выполнение второго постулата. При этом очевидно, что отличия должны становиться существенными только при очень больших скоростях, сравнимых со скоростью света $c\approx3\cdot10^8\;м/с$ (см. первую лекцию), поскольку окружающие нас повседневные явления классической теорией хорошо описываются. Прямые эксперименты с такими большими скоростями массивных объектов стали доступны далеко не сразу после появления СТО. Даже сейчас нет практической возможности сделать лекционные демонстрации, где проявились бы эффекты СТО. Поэтому, чтобы изучить следствия постулатов, обычно используются умозрительные эксперименты, из которых делаются логические выводы.

Чтобы уяснить суть отличий СТО от классической механики, можно начать со следующего простого эксперимента: пусть некторый вагон движется с постоянной скоростью мимо фонаря. В тот момент, когда фонарь поравнялся с серединой вагона, происходит его вспышка. В системе отсчета вагона свет распротраняется со скоростью $c$, и свет от вспышки достигает обоих концов вагона одновременно. Принимая во внимание второй постулат, в системе отсчета фонаря свет также движется со скоростью $c$, и сначала приходит в задний конец вагона, а затем - в передний. То есть, принятие второго постулата приводит к тому, что одновременные в одной инерциальной системе отсчета события (в данном случае - регистрация световых импульсов в противоположных концах вагона) оказываются не одновременными в другой инерциальной системе. Это - проявление относительности одновременности. В классической механике время считается абсолютным и текущим одинаково во всех инерциальных системах отсчета, и относительность одновременности не имеет места.

Будем далее называть событиями физические явления, происходящие в достаточно малой области пространства и длящиеся достаточно короткие промежутки времени, так что можно считать, что они происходят в какой-то фиксированной точке пространства в определенный момент времени (например, вспышка света и регистрация светового сигнала, как в предыдущем примере). Относительность одновременности приводит к необходимости рассматривать события в заданной инерциальной системе отсчета не только в пространстве, но и во времени.

### Изменение интервалов времени и длин

Рассмотрим простой умозрительный механизм - так называемые световые часы. Они представляют собой два зеркала, между которыми распространяется и отражается без потерь световой сигнал. Период прохождения сингалом расстояния между зеркалами $L$ "туда-обратно" задает фиксированный временной интервал, который можно принять за единицу измерения времени: $T_0 = 2L/c$.

Поместим световые часы на пару жестких рельс, так что часы могут свободно перемещаться вдоль оси $X$, и расстояние между рельсами по оси $Y$ фиксировано и равно $L$ (см. рисунок). Пусть $K$ - лабораторная система отсчета, относительно нее часы движутся вдоль рельса со скоростью $v$, и $K'$ - система, привязанная к часам. Найдем период часов в обеих ИСО, воспользовавшись постулатами Эйнштейна. В системе $K'$ период
$$ T' = T_0 = \dfrac{2L}{c} $$
В системе $K$ световой сигнал распространяется с той же скоростью $c$, но его точки отражения от зеркал смещаются в пространстве, поэтому, из теоремы Пифагора получаем
$$ (cT/2)^2 = L^2 + (vT/2)^2 \Rightarrow T = \dfrac{2L}{c}\dfrac{1}{\sqrt{1-v^2/c^2}}$$
Отсюда видно, что
$$ T = \dfrac{T_0}{\sqrt{1-v^2/c^2}} > T_0 $$
и можно сказать, что в системе отсчета, где часы покоятся, промежуток времени между двумя событиями - последовательными отражениями светового сигнала от зеркала - будет минимальным по сравнению с любой другой ИСО, где часы движутся. Введем так называемый гамма-фактор (или Лоренц-фактор):
$$\gamma = \dfrac{1}{\sqrt{1-v^2/c^2}},$$
причем при любой $v<c$ имеем $\gamma>1$. Тогда промежутки времени между двумя событиями в системах отсчета $K$ и $K'$ можно связать соотношением $T = \gamma T'$.

Пусть теперь есть пара часов, но движение системы $K'$ происходит вдоль направления распространения луча - вдоль оси $Y$. Соотношение периодов движущихся и неподвижных часов будет тем же самым, что и в предыдущем примере, поскольку каждые часы можно развернуть на $90^{\circ}$ в системе отсчета, где они покоятся, и их период при этом не поменяется. Для часов, движущихся в $K$ и неподвижных в $K'$ за счет смещения зеркал за время распространения светового луча имеем,
$$ T = T_1 + T_2, \; L + vT_1 = cT_1, \; L - vT_2 = cT_2 \Rightarrow T = \dfrac{L}{c-v} + \dfrac{L}{c+v} = \dfrac{2L}{c}\dfrac{1}{1-v^2/c^2}$$
Это - промежуток времени между последовательными отражениями от зеркала, измеренный в системе $K$. Подставим полученное выше соотношение между промежутками времени в разных ИСО и учтем в рассматриваемом случае продольного движения расстояние между зеркалами может зависеть от системы отсчета, в которой измеряется:
$$T = \dfrac{2L}{c}\dfrac{1}{1-v^2/c^2} = \dfrac{2L'}{c}\dfrac{1}{\sqrt{1-v^2/c^2}} \Rightarrow L = L'\sqrt{1-v^2/c^2}$$
Отсюда видно, что "видимый" продольный размер световых часов в системе отсчета $K$ оказывается меньшим размера $L'$ в системе $K'$, где часы покоятся. Более детальное обсуждение процесса измерений длин и того, что подразумевается под "видимым" размером, будет дано ниже.
