### Предмет общей физики
Данный курс носит название "Общей физики". В нем дается представление о широком круг явлений, известных современной физике, и при этом акцент делается на качественную интерпретацию, а не на строгое математическое описание физических моделей. С одной стороны общая физика подводит к более специализированным курсам математической и теоретической физики, а с другой - является самостоятельным предметом, формирующим у учащихся "физическую интуицию" и естественнонаучный кругозор.

Курс включает элементы следующих разделов современной физикии: механика, термодинамика и статистическая физика, электромагнетизм, оптика, квантовая физика, ядерная физика, астрофизика и физика конденсированного состояния. Первые четыре раздела составляют приблизительно содержание первых четырех семестров и являются фундаментом для всех последующих более специализированных курсов. Остальные упомянутые разделы затрагиваются в связи с обсуждением основополагающих идей и базовых экспериментальных фактов, их углубленное изучение будет происходить на старших курсах в зависимости от специализации учащихся.

### Научный подход

Наука развивалась на протяжении достаточно длительного времени, и сейчас можно сформулировать систему положений, на которые так или иначе опираются ученые в ходе своей научной деятельности. Несмотря на различие человеческих языков, людям удалось найти единый язык, удобный для описания природы - математику, - основы которой были заложены еще в древности [<a href="http://www.mathnet.ru/php/archive.phtml?wshow=paper&jrnid=ufn&paperid=2258&option_lang=rus">Птицы и лягушки в математике и физике</a>]. Достижения ученых древнего мира связаны в основном с геометрией и началами теори чисел. Математические основы движения планет, описанного Кеплером, лежат в евклидовой геометрии; теории электромагнетизма Максвелла - в теории дифференциальных уравнений; современной физики твердого тела - в топологии, теории групп и др. Это совершенно разные математические теории. Один из самых ярких примеров математического мышления в физике - общая теория относительности, для построения которой Эйнштейн использовал неевклидову геометрию, и отождествил физическое пространство-время с искривленным неевклидовым пространством. Подтверждения этой теории появились только после завершения ее создания. Квантовая механика также была создана на основе математических соображений: Шредингер заметил, что теория световых лучей представляет собой частный случай теории световых волн, и сделал предположение, что можно сформулировать аналогичную теорию волн по отношению к траекториям частиц.

Математика является основой для построения физических моделей, однако, полностью описать природу в рамках какой-то одной модели до сих пор не удалось. Поэтому процесс замены реальных явлений и физических объектов математическими абстрациями приводит к необходимости установления ограничений моделей. Задача физика - с одной стороны всесторонне изучить математические свойства моделей, а с другой - определить границы применимости моделей, то есть, условия, в которых правомерно эти модели применять для получения верных предсказаний.

Наука стремится к объективности и воспроизводимости, изучению и обобщению опыта всего человечества на протяжении длительного времени. При этом неявно предполагается, что мы верим в существование некоторой объективной реальности, общей для на всех нас и подчиняющейся одним и тем же законам, в любые моменты времени и во всех точках пространства. Целью научных исследований является выявление в совокупностях фактов общих закономерностей. Научное познание ориентировано на воспроизводимые во времени и пространстве результаты. Процесс исследований заключается в комбинации эмпирического (опытного) и теоретическо путей познания. Все начинается и заканчивается на опыте: опыт дает предпосылки для построения теоретических моделей, а предсказания последних, в свою очередь, роверяются на опыте.

### Системы мер и единиц

Обмен данными между учеными был бы сильно затруднен, если бы у них не было единой системы обозначений. Отголоски зоопарка единиц измерения, характерного для большей части Нового времени, можно видеть и сейчас в разнице различных систем мер и весов. Отдельные отрасли и профессии, например, морское или ювелирное дело, нефтедобыча, до сих пор оперируют своими специфическими единицами. Тем не менее, ситуация с обозначениями в корне изменилась после принятия в 1875 году Метрической конвенции и учреждения Международного бюро мер и весов. Специальный комитет решает задачу обеспечения существования единой системы измерений, вопросы стандартизации в области метрологии.

Кроме обеспечения существования универсальных единиц измерения, также необходимо свести эти единицы в системы, выделить *основные* и *производные*. Самой распространенной системой в мире является Международная система единиц СИ, утвержденная Генеральной конференцией по мерам и весам в 1960 году. Основными единицами системы СИ являются *килограмм, секунда, метр, кельвин, ампер, кандела и моль*. Они служат для измерения массы, времени, расстояния, термодинамической температуры, силы тока, силы света и количества вещества. Все остальные физические могут быть выражены через этот набор. В данном курсе мы встретимся и с другими системами, например, СГС, где основными являются *сантиметр, грамм, секунда*.

Анализ размерности физических величин в задачах играет очень важную роль. С одной стороны, получая какой-либо ответ, можно сразу же провести его проверку, проследив размерность результата. Если мы, к примеру, хотим найти ускорение, то, какое бы ни было сложное алгебраическое выражение в ответе, его размерность должна быть $м/с^2$. А величины, стоящие в аргументах математических функций, таких как $\sin(x)$ или $\exp(x)$, должны быть безразмерными. С другой стороны, иногда знание размерности параметров задачи и ответа, может натолкнуть на математическое выражение для ответа на задачу. К примеру, зададимся вопросом о засимости скорости свободно падающего тела от высоты падения, если тело отпущено без начальной скорости. Наш опыт (или задачник) говорит нам о том, что эта скорость может зависеть только высоты $h$ и ускорения свободного падения $g$. Их размерности $м$ и $м/с^2$ соответствнно. Наиболее простой комбинацией этих двух размерностей, позволяющей получить размерность скорости $м/с$, является $\sqrt{gh}$. Этот ответ оказывается правильным с точностью до постоянного безразмерного множителя $\sqrt{2}$, который невозможно получить в рамках подобных рассуждений. Более детально теория размерностей еще будет обуждаться в дальнейшем.

Недавним важным решением комитета мер и весов, вынесенным в 2019 году, была утверждена окончательная привязка всех основных единиц международной системы СИ к фундаментальным физическим константам, а также переопределен эталон килограмма (см., например, [<a href="https://www.nkj.ru/archive/articles/35677/">Посленим сдался килограмм</a>]). Необходимость такого решения давно назрела в связи с колоссальным увеличением точности физических измерений во второй половине прошлого века. Например, ранее единица измерения температуры была привязана к тройной точке воды, один кельвин равнялся $1/273.16$ температуры тройной точки воды. Проблема состоит в том, что незначительное изменение химического состава воды, может существенно сместить положение тройной точки, а, если привязаться к определенному составу, то возникает проблема его высокоточного контроля, так что возникают проблемы уже с контролем температуры на уровне долей милликельвинов. Еще хуже дело обстояло с одним килограммом. Его сравнивали с эталоном, сделанным во Франции в 1799 году, а затем замененным на новый эталон в 1889. В конце прошлого века ученые обнаружили, что массы разных копий эталона килограмма различаются на уровне десятков микрограммов, что является критическим для современных высокоточных измерений.

Чтобы устранить влияние условий проведения измерений, единицы были привязаны к фундаментальным константам, что отражает приведенная таблица [<a href="https://www.bipm.org/en/measurement-units">The International System of Units</a>]. Константы выбраны с учетом возможности их независимых измерений с высокой точностью. Это частота перехода между двумя сверхтонкими уровнями основного состояния изотопа атома цезия-133, скорость света, постоянная планка, элементарный заряд, постоянная Больцмана, постоянная Авогадро и постоянная максимальной световой эффективности. Численные значения, приведенные в таблице были зафиксированы из соображений соответствия с предыдущими менее точными определениями [<a href="https://ieeexplore.ieee.org/document/8716268">An introduction to the revised international system of units (SI)</a>].

| Основная единица СИ | Производная единица СИ | Фундаментальная константа | Зафиксированное значение | Формула связи |
| :- | :- | :- | :- | :- |
| $с$ | $Гц = с^{-1}$ | $\Delta\nu_{Cs}$ | 9192631770 | $\Delta\nu_{Cs}^{-1}$ |
| $м$ | $м/с$ | $c$ | 299792458 | $c/\Delta\nu_{Cs}$ |
| $кг$ | $Дж\cdot с = кг\cdot м^{2}\cdot с^{-1}$ | $h$ | $6.62607015\cdot 10^{-34}$ | $h\Delta\nu_{Cs}/c^2$ |
| $А$ | $Кл = А \cdot с$ | $e$ | $1.602176634\cdot 10^{-19}$ | $e\Delta\nu_{Cs}$ |
| $К$ | $Дж/К = кг\cdot м^{2}\cdot с^{-2}\cdot К^{-1}$ | $k_B$ | $1.380649\cdot 10^{-23}$ | $h\Delta\nu_{Cs}/k$ |
| $моль$ | $моль^{-1}$ | $N_A$ | $6.02214076\cdot 10^{23}$ | $1/N_A$ |
| $кд$ | $лм/Вт = кд\cdot ср/(кг\cdot м^{2}\cdot с^{-3})$ | $K_{cd}$ | 683 | $h(\Delta\nu_{Cs})^2/K_{cd}$ |

Таким образом, в отношении механических единиц,
* *секунда* - единица времени, равная $9\,192\,631\,770$ периодам излучения, соответствующего переходу между двумя сверхтонкими уровнями основного состояния атома цезия-133, находящегося в покое при 0 К. 
* *метр* - единица расстояния, равная расстоянию, которое проходит свет в вакууме за промежуток времени, равный $1/299792458$ секунды
* *килограмм* - единица массы, величина которой устанавливается фиксацией численного значения постоянной Планка $h$, равной в точности $6.62607015\cdot 10^{-34}$, когда она выражена единицей СИ Дж⋅с, которая эквивалентна $кг\cdot м^{2}\cdot с^{-1}$, где метр и секунда определены через $c$  и $\Delta\nu_{Cs}$.

### Масштабы современной физики

Необходимость описанных уточнений единиц измерения возникла из требований к точности современных фундаментальных экспериментов. Современная физика оперирует очень разными масштабами, а шкала расстояний, доступных для экспериментальных измерений сильно изменилась на протяжении последних веков. Приведем несколько примеров:

Масштабы микромира:
* размеры атомов и молекул: $\sim10^{-10}-10^{-6}$ м (от ангстремов до микрометров)
* электронные орбиты в атоме: $\sim10^{-10}$ м (ангстрем)
* размеры атомных ядер: $\sim10^{-14}-10^{-13}$ м
* размеры нуклонов: $\sim10^{-15}$ м (фемтометр)

Космические и космологические масштабы:
* Расстояние от Солнца до Нептуна: $\approx4.5\cdot10^{9}$ км
* Диаметр галактики $\sim10^{5}$ световых лет ($\sim10^{16}$ м)
* Размер известной области Вселенной: $\sim4.6\cdot10^{10}$ световых лет

Для высокоточных измерений требуются высокоточные приборы. Мощный толчок развитию технологий в совершенно разных областях дало изобретение лазера в пятидесятые годы 20-го столетия. Так, гравитационные волны, предсказанные в 1916 году А. Эйнштейном, были открыты в 2016 году с помощью по сути с помощью оптического прибора - лазерного интерферометра Майкельсона [<a href="https://habr.com/ru/post/407499/">LIGO: линейка точностью в 1/10000 диаметра протона</a>]. Принцип действия этого прибора основан на явлении интерференции. Сначала лазерный луч расщепляется на два одинаковых луча, которые после прохождения некоторого расстояния в плечах интерферометра вновь встречаются и складываются на фотодетекторе. Ток фотодетектора меняется в зависимости от разности фаз пришедших пучков, а изменение разности фаз напрямую связано с разностью пройденных расстояний. Более продробно различные схемы интерферометров будут обсуждаться в разделе, посвященном классической оптике. В детекторе LIGO достигается чуствитиельность $10^{-21}$ м. Подобный детектор способен фиксировать движение велосипедиста на расстоянии километра, или шум прибоя в тысячах километров от установки.

Другой пример - применение существенно квантовых явлений для измерения сверхмалых магнитных полей. Для этого применяется так называемый СКВИД - сверхпроводящий квантовый интерферометр. Здесь тоже используется явление интерференции, но не световых лучей, а волновых функций электронов, являющихся носителями тока в электрической цепи прибора. Наличие магнитного поля приводит к появлению дополнительного тока в витке, что приведет к возникновению разности фаз для двух участков витка, так что ток прибора будет зависеть от приложенного магнитного поля. Таким образом можно измерять магнитные поля $10^{-17}$ Тл (для сравнения магнитное поле Земли $\sim10^{-4}$ Тл)

Квантовые явления позволили и переопределить килограмм. Теперь зафиксированная постоянная планка позволяет определять массу с помощью отработанной процедуры - взвешивании на весах Киббла. Принцип работы таких весов состоит в уравновешивании массы с помощью силы Ампера, действующей со стороны магнитно поля на проводник с током. На следующем этапе, чтобы избежать измерения магнитно поля, измеряется напряжение на концах проводника, возникающее вследствие явления электромагнитной индукции. Таким образом, масса становится пропорциональной произведению напряжения и тока:
$$mg=BLI,\;\;U=BLv\;\Rightarrow\;UI=mgv$$

При этом оказывается, что $UI\sim h$. Отсюда вытекает связь массы и постоянной Планка.

### Кому еще нужна физика?

Современная физика включает в себя огромное можество областей, большинство из которых тесно переплетены с другими наукамми - химией, билогией, материаловедением, и даже экономикой.

Приведем пример простейшего физического анализа параметров дорожного движения. По аналогии с интутивно ясным понятием потока жидкости - массы, протекающей через заданное сечение в единицу времени, - можно ввести понятие потока для совершенно разных величин. В дальнейшем мы будем работать, например, с потоками импульса и энергии. Можно рассмотреть и поток автомобилей на трассе. Если $\rho$ - число автомобилей на единицу длины дороги (линейная плотность), а $v$ - средняя скорость движения, то поток
$$\Phi = \rho v$$
Число автомобилей на единицу длины дороги можно выразить черз среднюю длину автомобиля $l$ и среднее расстояние между транспортными средствами $L$:
$$\rho = \frac{1}{l+L}$$
Согласно правилам дорожного движения, рекомендуемая дистанция между транспортными средствами в метрах равна половине скорости автомобиля в км/ч, то есть, $L = \alpha v$, где $\alpha = 0.5 м/(км/ч) = 1.8 с$. Тогда поток, или, другими словами, пропускная способность одного ряда в секунду
$$\Phi = \frac{v}{l+\alpha v}\rightarrow0.56 с^{-1},\;v\rightarrow\infty$$
Замечательно, что при увеличении скорости до бесконечности, поток стремится к постоянной фиксированной величне, так что за минуту по полосе будет проезжать не более 33 автомобилей. Это, конечно, при условии строгого соблюдения правил. В противном случае, если же скрость роста функции $L(v)$ медленее линейной, поток будет стремиться к бесконечности.

В этом семестре мы будем изучать основы классической механики. Слово "классическая" подразумевает, что изучаемые законы будут касаться макромира и объектов, состоящих из очень большого числа частиц, так что проявлениями квантовых эффектов можно пренебречь. Будут рассмотрены как основы нерелятивистской, так и релятивитсткой теории (специальной теории относительности). В первом случае предполагается, что все движения происходят со скростями много меньшими скорости света в вакууме.
