### Синхронизация часов

В примере на предыдущей лекции неявно предполагалось, что в разных пространственных точках одной и той же ИСО часы идут одинаково. С учетом того, что скорость распространения света и всяких информационных сигналов конечна, перед сравнением часов в разных системах отсчета необходимо предусмотреть синхронизацию часов в одной ИСО. Можно представить, что в каждой пространственной точке системы отсчета расположены часы. Достаточно рассмотреть процедуру синхронизации только для двух точек, тогда ее можно будет провести для любого другого числа точек, где необходимо расположить часы.

Можно рассмотреть различные процедуры выставления одного и того же времени на часах одной ИСО. Пусть синхронизируемые часы расположены в точках $A$ и $B$ оси $X$. Выпустим из точки $A$ световой сигнал в момент времени $t_1$ в направлении точки $B$. Если в точке $B$ установить зеркало, то сигнал отразится и вернется в точку $A$ в некоторый момент $t_2$ по часам $A$. Поскольку время распространения сигнала туда и обратно одно и то же, часы в точке $B$ нужно выставить так, чтобы в момент отражения сигнала их показания былы равны $(t_1+t_2)/2$. Можно рассмотреть и другую процедуру: неподвижной в данной ИСО линейкой отмерить середину расстояния между $A$ и $B$, испустить одновременно из этой точки световые сигналы в $A$ и $B$, и зафиксировать одно и то же время на часах $A$ и $B$ в момент приема сигналов.

### Вывод преобразования Лоренца

Преобразования координат и времени между разными инерциальными системами отсчета (ИСО) называются преобразованиями Лоренца (в честь нидерландского физика <a href="https://indicator.ru/physics/nobelevskie-laureaty-hendrik-lorenc.htm">Хендрика Лоренца</a>). Эти преобразования можно вывести разными способами. Здесь мы рассмотрим два подхода. Далее мы будем рассматривать пары движущихся относительно друг друа ИСО. Без ограничения общности можно считать, что движение происходит вдоль одной из декартовых координатных осей в обоих системах, например осей $X$.

Рассмотрим "лабораторную" ИСО $K$ и ИСО $K'$, движущуюся относитльно нее с постоянной скоростью $v$. Пусть в моменты времени $t=0$ и $t'=0$, измеренные в этих системах, начала систем $O$ и $O'$ совпадают. Часы, находящиеся в этих точках, будем называть реперными. Пространственные и временные координаты, соответствующие рассматриваемым ИСО, будем обозначать нештрихованными и штрихованными величинами соответственно. Зададимся вопросом, что будут показывать часы в разных системах отсчета, оказавшиеся в одной точке пространства? Рассмотренное выше явление изменения расстояния между зеркалами световых часов можно обобщить, сказав, что координатная сетка системы $K'$ относительно наблюдателя из системы $K$ сжата в $1/\gamma$ раз. Тогда в начальный момент времени координаты в рассматриваемых системах связаны соотношением $x(0) = x'/\gamma$. К моменту времени $t$ вся сетка в $K'$ сместится на расстояние $vt$, поэтому
$$x = x'/\gamma + vt \Rightarrow x' = \gamma(x-vt)$$

Определим положение сравниваемых часов в двух системах с помощью светового сигнала: испустим в начальный момент времени по часам $K$ и $K'$ из начала координат световой сигнал. В момент $t$ сигнал приходит в точку $x_2=ct$. Координаты этого же события в $K'$ связаны соотношением $x_2'=ct_2'$. Согласно найденному закону преобразования координат,
$$\gamma(x_2-vt)=ct_2' \Rightarrow t_2' = \gamma t(1-\beta)$$
где $\beta = v/c$. В момент $t$ начало системы $K'$ сместится в точку $x_1=vt$ - там будут находится опорные часы системы $K'$. Время, которое они показывают, $t_1'=t/\gamma$ (согласно рассмотренному выше эксперименту со световыми часами). Тогда в один и тот же момент времени $t$ по часам системы $K$ часы из $K'$, оказавшиеся в разных точках, покажут разное время:
$$x_{1}=Vt\Rightarrow t_{1}'=t/\gamma$$
$$x_{2}=ct\Rightarrow t_{2}'=\gamma t\left(1-\beta\right)$$
При этом зависимость рассинхронизации часов от смещения вдоль оси $X$ оказывается линейная:
$$\frac{\Delta t'}{\Delta x} = \frac{t_2'-t_1'}{x_2-x_1} = \frac{\gamma t(1-\beta) - t/\gamma}{ct-vt} = -\frac{\gamma\beta}{c}$$

Полученное выше соотношение $x' = \gamma(x-vt)$ задает преобразование пространственной координаты. Чтобы получить преобразование временной координаты, нужно выразить показания часов из $K'$, которые в момент времени $t$ по часам $K$ оказались в точке $x$. Эти часы в начальный момент находились в точке $x-vt$, а, следовательно, отставали от опорных часов в $K'$ на промежуток $\Delta t' = -\gamma\beta(x-vt)/c$. В момент $t$ эти опорные часы показывают время $t/\gamma$, следовательно, искомые показания часов в точке $x$ будут
$$t'=t/\gamma+\Delta t'=t/\gamma-\dfrac{\gamma\beta}{c}\left(x-Vt\right)=\gamma\left(t-\dfrac{\beta}{c}x\right)$$

### Другой вывод преобразования Лоренца

К полученным преобразованиям можно прийти и другим способом. Здесь также будем рассматривать две ИСО, $K$ и $K'$, так что $K'$ движется относительно $K$ со скоростью $v$ вдоль оси $X$ в положительном направлении. Все опытные данные говорят нам о том, что пространство и время однородны. В силу этого любой линейный сдвиг пространственных и временных координат не должен приводить к изменению физических законов. При использовании линейного преобразования между различными ИСО:
$$x'=a_{1}x + b_{1}t + c_{1}$$
$$t'=a_{2}x + b_{2}t + c_{2}$$
линейный сдвиг $x\rightarrow x+\xi$, $t\rightarrow t+\tau$ дает аналогичный линейный сдвиг и в любой другой системе отсчета:
$$x'\rightarrow x' + a_{1}\xi + b_{1}\tau$$
$$t'\rightarrow t' + a_{2}\xi + b_{2}\tau$$
Поэтому из однородности пространства следует линейность преобразований. Выберем начало отсчета таким образом, чтобы прямое и обратное преобразования можно было записать
$$x' = a_{1}x + b_{1}t$$
$$x = \tilde{a}_{1}x' + \tilde{b}_{1}t'$$
Координата начала отсчета системы $K'$ в системе $K$ есть $x_{O'}=vt$, а в системе $K'$
$$x'_{O'}\equiv 0 \Rightarrow 0 = a_{1}vt + b_{1}t \Rightarrow \frac{b_{1}}{a_{1}}=-v$$
Аналогично можно сделать вывод, что $\tilde{b}_{1}/\tilde{a}_{1}=v$, откуда
$$x' = a_{1}(x - vt)$$
$$x = \tilde{a}_{1}(x' + vt')$$
Поскольку любые ИСО равноправны, должно выполняться равенство $a_{1}=\tilde{a}_{1}$. Для определения этой постоянной, рассмотрим световой сигнал, испущенный в начальный момент времени из начал рассматриваемых систем. Координаты этого сигнала $x=vt$ и $x'=vt'$ соответственно. Подставляя эти соотношения, имеем,
$$x' = ct' = a_{1}(ct - vt)$$
$$x = ct = a_{1}(ct' + vt')$$
$$\Rightarrow a_{1} = \frac{1}{\sqrt{1-v^2/c^2}} = \gamma$$

### Преобразования Лоренца (резюме)

Резюмируя полученные выше преобразования, в случае, когда система $K'$ движется относительно системы $K$ в положительном направлении относительно оси $X$ со скоростью $v$,
$$x' = \gamma(x-\beta\tau)$$
$$\tau' = \gamma(\tau-\beta x)$$
$$y'=y,\;z'=z$$
где введено обозначение $\tau=ct$. Обратное преобразование получается заменой знака у относительной скорости систем отсчета:
$$x = \gamma(x'+\beta\tau')$$
$$\tau = \gamma(\tau'+\beta x')$$
Если рассматриваются два события, координаты которых в пространстве и времени $(x_1,\tau_1)$ и $(x_2,\tau_2)$ в ИСО $K$, и $(x_1',\tau_1')$ и $(x_2',\tau_2')$ в ИСО $K$, то изменения пространственно-временных координат между этими событиями ($\Delta x = x_2-x_1$, $\Delta \tau = \tau_2-\tau_1$, $\Delta x' = x_2'-x_1'$, и $\Delta \tau' = \tau_2'-\tau_1'$) будут связаны с помощью преобразований Лоренца:
$$\Delta x' = \gamma(\Delta x-\beta\Delta \tau)$$
$$\Delta \tau' = \gamma(\Delta \tau-\beta \Delta x)$$
$$\Delta x = \gamma(\Delta x'+\beta\Delta\tau')$$
$$\Delta\tau = \gamma(\Delta\tau'+\beta \Delta x')$$

### Собственное время

Время, отсчитываемое по покоящимся часам (привязанным к некоторому физическому объекту), называется собственным временем. При рассмотрении опыта со световыми часами, было найдено, что время между двумя событиями, отсчитываемое по таким часам, всегда меньше времени между теми же событиями, измеренному в любой другой ИСО, где эти часы движутся. Связь между промежутками можно получить непосредственно из преобразований Лоренца: пусть ИСО $K'$ движется относительно ИСО $K$ со скоростью ИСО $v$. Тогда время, отсчитываемое по покоящимся в $K'$ часам ($\Delta x'=0$) будет собственным, $t_c = \Delta t'$. Согласно преобразованию времени
$$\Delta t = \gamma (\Delta t' + (\beta/c)\Delta x') = \gamma t_c$$
Подобное соотношение часто называеют эффектом замедления времени.

Для иллюстрации проявления эффекта, достоверно обнаруженного экспериментально, часто приводят пример регистрации частиц - мюонов - в земной атмосфере. В космическом пространстве существует непрерывный поток первичного космического излучения - поток выскоэнергетичных атомных ядер. Эти ядра, сталкиваясь с ядрами атомов, составляющих атмосферу Земли, приводят к частичному расщеплению последних, а также рождению новых частиц, в частности пионов $\pi^{\pm}$. Время жизни пионов (то есть, характерное время, за которое они превращаются в другие частицы) в их системе отсчета очень мало - примерно, $0.8\cdot10^{-16}$с. При распаде пионов образуются мюоны и нейтрино: $\pi^{\pm}\rightarrow\mu^{\pm}+\nu_{\mu}(\tilde{\nu}_{\mu})$. Другие продукты подобных превращений - в основном, электроны и фотоны - сильно поглощаются в атмосфере, так что на поверхности Земли в лабораторных условиях наблюдаются преимущественно мюоны, появившиеся таким образом. Характерная скорость мюонов $v\sim0.998c$. Время жизни мюона (собственное время) - $t_c = 2\cdot10^{-6}$с. Если бы не существовало эффекта замедления времени характерный пробег мюонов был бы $l=vt_c\approx600$ м. И регистрация у поверхности Земли была бы невозможна. Если же пересчитать пробег в системе отсчета земного наблюдателя с помощью преобразования Лоренца (пространственный промежуток между событиями рождения мюона и его регистрацией), получаем
$$l=\gamma v\Delta t' = \gamma v t_c \approx 10\;км$$
что примерно соответствует границе тропосферы. Поток мюонов таков, что на один квадратный метр попадает порядка $10^4$ частиц в минуту.

### Измерение длин

Измерение длины в ИСО, где измеряемый объект и линейка покоятся, можно осуществить просто прикладыванием линейки. Подобный процесс невозможен в ситуации движущегося объекта. Можно представить себе наблюдателя, относительно которого движется поезд. Поезд проезжает мимо часто расставленных столбиков, так что наблюдатель пытается заметить, радом с каким из столбиков находится начало или конец вагона. Если наблюдатель заметил положение начала вагона, то за время поворота головы, поезд успеет сместиться, так что наблюдаемое положение конца вагона не будет соответствовать его положению в момент наблюдения положения начала. Так что наблюдатель должен в итоге учесть скорость поезда, чтобы получить правильную длину по своим столбикам. Эта иллюстрация не совсем корректна с точки зрения релятивистской теории, тем не менее, она демонстрирует необходимость уточнения процесса измерения длины. И, так назваемый эффект сокращения длины возникает именно из-за того, что в разных системах отсчета измерение длин дает разный результат.

Пусть в ИСО $K$ движется протяженный объект со скоростью $v$, и пусть длина объекта в системе отсчета $K'$, где он покоится, равна $l_0$. Первый способ измерить его длину - зафиксировать положения начала и конца объекта в один и тот же момент времени по часам системы $K$. То есть, подразумевается, что в каждой точке оси $X$ системы $K$ сидят наблюдатели, их часы синхронизированы, и в определенный момент времени двое из этих наблюателей видят начало и конец объекта. Измеренную таким образом длину можно получить из преобразования Лоренца, учитывая, что мы ищем $\Delta x = l$, собственная длина стержня $\Delta x'=l_0$, промежуток времени между измерениями в $K$ $\Delta t=0$, а в системе $K'$ этот промежуток нам не важен:
$$\Delta x' = \gamma(\Delta x + v\Delta t) \Rightarrow l = l_0/\gamma = l_0\sqrt{1-\beta^2}$$
Это тот же результат, что и был получен при рассмотрении световых часов.

Второй способ, требующий наличия всего одного наблюдателя в системе $K$ - отметить промежуток времени $\Delta t$ между моментами пролета концов стержня мимо этого наблюдателя. Тогда рассматривая эти два события в системах наблюдателя и стержня, имеем, $\Delta x = 0$, $\Delta t = l/v$, $\Delta x' = -l_0$. Следовательно, преобразование Лоренца даст
$$\Delta x' = \gamma(\Delta x - v\Delta t) \Rightarrow l = l_0/\gamma$$
то есть, тот же самый результат.

Рассмотрим еще один способ измерения длины: пусть в системе отсчета $K'$, связанной с летящим объектом, на концах объекта одновременно вспыхивают лампочки ($\Delta t'=0$). Зафиксированное расстояние между лампочками в системе $K$ можно найти из преобразования Лоренца: $\Delta x = \gamma\Delta x' = \gamma l_0$. Но это расстояние не будет измеряемой длиной, поскольку в $K$ лампочки вспыхивают не одновременно, и наблюдатель из $K$ должен сделать поправку на смещение объекта за время между вспышками. Это время равно $\Delta t=\gamma(\beta/c)\Delta x' = \gamma l_0v/c^2$. Тогда измеряемая длина
$$l = \gamma l_0 - v\Delta t = \gamma l_0 - v\gamma l_0v/c^2 = l_0/\gamma$$

### Интервал

При рассмотрении переходом между различными ИСО важными являются величины, инвариантные относительно преобразований координат. Например, при преобразованиях Галилея инвариантами являлись расстояния между точками пространства (или длина физических объектов) и промежутки времени. Из вышеприведенной теории следует, что эти величины уже не будут инвариантны относительно преобразований Лоренца. В релятивистском случае инвариантом оказывается специальная комбинация пространственного расстояния и промежутка времени между двумя событиями, называемая интервалом:
$$\Delta s = \sqrt{(c\Delta t)^2 - (\Delta x)^2}$$
Покажем, что интервал инвариантен относительно преобразования Лоренца:
$$(\Delta s')^2 = (c\Delta t')^2 - (\Delta x')^2 = \left(c\dfrac{\Delta t-\left(V/c^{2}\right)\Delta x}{\sqrt{1-V^{2}/c^{2}}}\right)^{2}-\left(\dfrac{\Delta x-V\Delta t}{\sqrt{1-V^{2}/c^{2}}}\right)^{2} = $$
$$=\dfrac{\left(c\Delta t\right)^{2}+\left(V/c\right)^{2}\left(\Delta x\right)^{2}-2V\Delta x\Delta t}{1-V^{2}/c^{2}}-\dfrac{\left(\Delta x\right)^{2}-V^{2}/c^{2}\left(c\Delta t\right)^{2}-2V\Delta x\Delta t}{1-V^{2}/c^{2}} = $$
$$=\dfrac{\left(c\Delta t\right)^{2}\left(1-V^{2}/c^{2}\right)-\left(\Delta x\right)^{2}\left(1-V^{2}/c^{2}\right)}{1-V^{2}/c^{2}}=\left(\Delta s\right)^{2}$$
Необходимо отметить, что существует возможность вывести инвариантность интервала непосредственно из постулатов Эйнштейна, и на этом основании получить явный вид преобразований Лоренца (см., например, ...).

С помощью интервала можно сформулировать условия, при которых существуют системы отсчета, в которых пары событий происходят в одной точке пространства, или в один и тот же момент времени. Если есть инерциальная система отсчета, в которой пара событий произошла в одной точке пространства, то для них $x_1=x_2$, и $(\Delta s)^2 = c^2(\Delta t)^2 > 0$. Ввиду инвариантности, кварат интервала между этими событиями будет положительным и в любой другой ИСО, а сам интервал будет вещественным. Вещественные интервалы называются времениподобными. Аналогично рассмотрение событий, проиходящих в один и тот же момент времени в некоторой ИСО, приводит к тому, что интервал между ними оказыватся мнимым, $\Delta s = \sqrt{-(\Delta x)^2}$, и такие интервалы называются пространственноподобными.

Рассмотрим пару событий $(x_1, t_1)$ и $(x_2, t_2)$ в некоторой ИСО. Если первое событие стало причиной второго, значит, имело место некоторое физическое взаимодействие. Пусть скорость передачи этого взаимодействия равна $v_0$, причем $v_0\leq c$, поскольку скорость света является предельной скоростью распространения взаимодействий, и
$$t_2-t_1 = \frac{x_2-x_1}{v_0} > 0$$
Покажем, что преобразования Лоренца сохраняют причинно-следственную связь, то есть, в любой системе отсчета первое событие будет причиной второго. В произвольной ИСО, движущейся относительно исходной со скоростью $v$, промежуток времени между событиями
$$t_{2}'-t_{1}'=\dfrac{t_{2}-t_{1}-\left(v/c^{2}\right)\left(x_{2}-x_{1}\right)}{\sqrt{1-v^{2}/c^{2}}}=\dfrac{t_{2}-t_{1}}{\sqrt{1-v^{2}/c^{2}}}\left(1-\dfrac{v_{0}v}{c^{2}}\right)\geq0$$
Причинность подразумевает, что $\Delta x\leq c\Delta t$, значит, интервал между связанными событиями должен быть времениподобным.
